## 3- nettoyer les colonne de 'Prix'

```python
# remplacer les serie de nan if 'PRIX NON SPECIFIE' kina
df.loc[df['Prix'] == 'PRIX NON SPÉCIFIÉ', 'Prix'] = np.nan

# supprimer les str DH Est '/Nuit' dans la colonne de "Pix"
df.loc[:, 'Prix'] = df['Prix'].replace(r'\D', '', regex=True)
# convertir la colonne en type numerique 
df['Prix'] = pd.to_numeric(df['Prix'], errors='coerce')
```

## salons :

## prix 

```python
# Convertir la colonne 'Prix' en numérique
df['Prix'] = pd.to_numeric(df['Prix'], errors='coerce')

# Calcul des statistiques descriptives
moyenne = df['Prix'].mean()
mediane = df['Prix'].median()

# Vérification de la distribution
if abs(moyenne - mediane) > 0.1 * moyenne:
    print("Distribution asymétrique")
else:
    print("Distribution symétrique")
```

```python
# Imputer les valeurs manquantes avec la médiane
df['Prix'] = df['Prix'].fillna(df['Prix'].median())
```

## surface :

```python
# mean est median de colonne Surfece_habitable .
mean_surface = df['Surface_habitable'].mean()
median_surface = df['Surface_habitable'].median()

print(f"Mean of Surface_habitable: {mean_surface}")
print(f"Median of Surface_habitable: {median_surface}")
```

```python
# charger les valeur nan en mean ,
df['Surface_habitable'] = df['Surface_habitable'].fillna(mean_surface)
```

```python
# nest pas une valeur null
print(df['Surface_habitable'].isnull().sum())
```

## etage :

```python
# Trouver la valeur la plus fréquente dans la colonne 'Étage'
mode_etage = df['Étage'].mode()[0]

# Remplir les valeurs nulles dans la colonne 'Étage' avec la valeur la plus fréquente
df['Étage'] = df['Étage'].fillna(mode_etage)
```

## Âge_du_bien

```python
# Trouver la valeur la plus fréquente dans la colonne 'Âge_du_bien'
mode_age = df['Âge_du_bien'].mode()[0]

# Remplir les valeurs nulles dans la colonne 'Âge_du_bien' avec la valeur la plus fréquente
df['Âge_du_bien'] = df['Âge_du_bien'].fillna(mode_age)
```
